Problem Statement.

You are given a string s. You can convert s to a by adding characters in front of it.

Return the shortest palindrome you can find by performing this transformation.

 

Example 1:

Input: s = "aacecaaa"
Output: "aaacecaaa"

Example 2:

Input: s = "abcd"
Output: "dcbabcd"

 

Constraints:

    0 <= s.length <= 5 * 104
    s consists of lowercase English letters only.


# KNP Algorithm - O(N) runtime, O(N) space

```python
class Solution:
    def shortestPalindrome(self, s: str) -> str:
        reversed_string = s[::-1]
        combined_string = s + "#" + reversed_string
        prefix_table = [0] * len(combined_string)
        length = 0

        for i in range(1, len(combined_string)):
            while length > 0 and combined_string[i] != combined_string[length]:
                length = prefix_table[length - 1]
            if combined_string[i] == combined_string[length]:
                length += 1
            prefix_table[i] = length

        palindrome_length = prefix_table[-1]
        suffix = reversed_string[: len(s) - palindrome_length]
        return suffix + s
```

```python
instance = Solution()
instance.shortestPalindrome("abcd")
```
