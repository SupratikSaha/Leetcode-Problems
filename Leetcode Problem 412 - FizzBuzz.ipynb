```python
class Solution:
    def fizzBuzz(self, n: int):
        
        if not isinstance(n,int) or n <= 0:
            return []
        
        res = []
        for idx in range(1,n+1):
            if idx%3 != 0 and idx%5 != 0:
                res.append(str(idx))
            elif idx%3 == 0 and idx%5 == 0:
                res.append("FizzBuzz")
            elif idx%3 == 0:
                res.append("Fizz")
            else:
                res.append("Buzz")
                
        return res
```

```python
Instance = Solution()
Instance.fizzBuzz(15)
```
