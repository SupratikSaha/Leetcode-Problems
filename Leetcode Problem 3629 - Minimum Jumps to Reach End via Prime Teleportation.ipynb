Problem Statement.

You are given an integer array nums of length n.

You start at index 0, and your goal is to reach index n - 1.

From any index i, you may perform one of the following operations:

    Adjacent Step: Jump to index i + 1 or i - 1, if the index is within bounds.
    Prime Teleportation: If nums[i] is a p, you may instantly jump to any index j != i such that nums[j] % p == 0.

Return the minimum number of jumps required to reach index n - 1.

 

Example 1:

Input: nums = [1,2,4,6]

Output: 2

Explanation:

One optimal sequence of jumps is:

    Start at index i = 0. Take an adjacent step to index 1.
    At index i = 1, nums[1] = 2 is a prime number. Therefore, we teleport to index i = 3 as nums[3] = 6 is divisible by 2.

Thus, the answer is 2.

Example 2:

Input: nums = [2,3,4,7,9]

Output: 2

Explanation:

One optimal sequence of jumps is:

    Start at index i = 0. Take an adjacent step to index i = 1.
    At index i = 1, nums[1] = 3 is a prime number. Therefore, we teleport to index i = 4 since nums[4] = 9 is divisible by 3.

Thus, the answer is 2.

Example 3:

Input: nums = [4,6,5,8]

Output: 3

Explanation:

    Since no teleportation is possible, we move through 0 → 1 → 2 → 3. Thus, the answer is 3.

 

Constraints:

    1 <= n == nums.length <= 105
    1 <= nums[i] <= 106


# BFS - O(M log log M + N log M) runtime, O(M + N + N * Log M) space, where M is Max number

```python
from typing import List
from collections import defaultdict, deque
from math import ceil, sqrt

class Solution:
    def minJumps(self, nums: List[int]) -> int:
        i, n, maxi = 2, len(nums), max(nums) + 1
        dic = defaultdict(list)
        prime = [i for i in range(maxi)]

        #Finding Primes
        for p in range(i*i, maxi, i):
            prime[p]=i
        for i in range(3, ceil(sqrt(maxi)), 2):
            if prime[i] == i:
                for p in range(i*i, maxi, i):
                    prime[p] = i

        #Finding Indices divisible by each Primes
        for i in range(n):
            num = nums[i]
            while num != prime[num]:
                prnum = prime[num]
                dic[prime[num]].append(i)
                while num % prnum == 0:
                    num //= prnum
            if num != 1:
                dic[num].append(i)
        
        jumpedPrimes = set()
        q = deque([(0,0)])
        vis = [False]*n
        vis[0] = True

        while q:
            i, steps = q.popleft()
            if i == n-1: return steps
            if i+1 < n and not vis[i+1]:
                vis[i+1] = True
                q.append((i+1, steps+1))
            if i-1 >= 0 and not vis[i-1]:
                vis[i-1] = True
                q.append((i-1, steps+1))
            if prime[nums[i]] == nums[i] and nums[i] not in jumpedPrimes:
                jumpedPrimes.add(nums[i])
                for j in dic[nums[i]]:
                    if not vis[j]:
                        vis[j] = True
                        q.append((j, steps+1))
```

```python
instance = Solution()
instance.minJumps([1,2,4,6])
```
