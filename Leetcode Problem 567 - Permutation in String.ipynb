Problem Statement.

Given two strings s1 and s2, return true if s2 contains a of s1, or false otherwise.

In other words, return true if one of s1's permutations is the substring of s2.

 

Example 1:

Input: s1 = "ab", s2 = "eidbaooo"
Output: true
Explanation: s2 contains one permutation of s1 ("ba").

Example 2:

Input: s1 = "ab", s2 = "eidboaoo"
Output: false

 

Constraints:

    1 <= s1.length, s2.length <= 104
    s1 and s2 consist of lowercase English letters.

# Sliding Window - O(S + T), O(S + T) space

```python
from collections import Counter, defaultdict

class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        m, n = len(s1), len(s2)
        if m > n: return False
        chars_matched = 0
        left, char_dict = 0, defaultdict(int)
        ctr = Counter(s1)

        for right, c in enumerate(s2):
            if c not in ctr:
                left, chars_matched, char_dict = right + 1, 0, defaultdict(int)
            else:
                char_dict[c] += 1
                if char_dict[c] == ctr[c]:
                    chars_matched += 1
                    if chars_matched == len(ctr): return True
                while char_dict[c] > ctr[c]:
                    if char_dict[s2[left]] == ctr[s2[left]]:
                        chars_matched -= 1
                    char_dict[s2[left]] -= 1
                    left += 1

        return False 
```

```python
instance = Solution()
instance.checkInclusion("ab", "eidbaooo")
```
