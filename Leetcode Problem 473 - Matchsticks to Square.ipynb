Problem Statement.

You are given an integer array matchsticks where matchsticks[i] is the length of the ith matchstick. You want to use all the matchsticks to make one square. You should not break any stick, but you can link them up, and each matchstick must be used exactly one time.

Return true if you can make this square and false otherwise.

 

Example 1:

Input: matchsticks = [1,1,2,2,2]
Output: true
Explanation: You can form a square with length 2, one side of the square came two sticks with length 1.

Example 2:

Input: matchsticks = [3,3,3,3,4]
Output: false
Explanation: You cannot find a way to form a square with all the matchsticks.

 

Constraints:

    1 <= matchsticks.length <= 15
    1 <= matchsticks[i] <= 108



# DP - O(N * 2 ^ N) runtime, O(N + 2 ^ N) space

```python
from typing import List

class Solution:
    def makesquare(self, matchsticks: List[int]) -> bool:
        n = len(matchsticks)
        perimeter = sum(matchsticks)
        side_len, extra =  divmod(perimeter, 4)
        if n < 4 or extra != 0: return False

        def recurse(mask, sides_done):
            cur_used = 0
            for i in range(n):
                if not (mask & (1 << i)): cur_used += matchsticks[i]

            if cur_used > 0 and cur_used % side_len == 0: sides_done += 1
            if sides_done == 3: return True
            if (mask, sides_done) in memo: return memo[(mask, sides_done)]
            res = False
            rem = side_len - cur_used % side_len
            rem = side_len if rem == 0 else rem

            for i in range(n):
                if matchsticks[i] <= rem and mask & (1 << i):
                    if recurse(mask ^ (1 << i), sides_done):
                        res = True
                        break
                
            memo[(mask, sides_done)] = res
            return res

        memo = {}
        return recurse((1 << n) - 1, 0)
```

```python
instance = Solution()
instance.makesquare([1,1,2,2,2])
```
