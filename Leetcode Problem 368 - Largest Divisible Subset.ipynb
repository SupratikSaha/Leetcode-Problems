Problem Statement.

Given a set of distinct positive integers nums, return the largest subset answer such that every pair (answer[i], answer[j]) of elements in this subset satisfies:

    answer[i] % answer[j] == 0, or
    answer[j] % answer[i] == 0

If there are multiple solutions, return any of them.

 

Example 1:

Input: nums = [1,2,3]
Output: [1,2]
Explanation: [1,3] is also accepted.

Example 2:

Input: nums = [1,2,4,8]
Output: [1,2,4,8]

 

Constraints:

    1 <= nums.length <= 1000
    1 <= nums[i] <= 2 * 109
    All the integers in nums are unique.

# Hashmap - O(N ^ 2) runtime, O(N ^ 2) space

```python
from typing import List

class Solution:

    def largestDivisibleSubset(self, nums: List[int]) -> List[int]:
        subsets = {-1: set()}

        for num in sorted(nums):
            subsets[num] = max([subsets[k] for k in subsets if num % k == 0], key=len) | {num}

        return list(max(subsets.values(), key=len))
```

# DP - O(N ^ 2) runtime, O(N) space

```python
from typing import List

class Solution:
    def largestDivisibleSubset(self, nums: List[int]) -> List[int]:
        nums.sort()
        dp = [0] * len(nums)

        for i, num in enumerate(nums):
            max_size = 0
            for k in range(i):
                if nums[i] % nums[k] == 0:
                    max_size = max(max_size, dp[k])

            max_size += 1
            dp[i] = max_size

        max_len, idx = max([(v, i) for i, v in enumerate(dp)])
        res = []
        cur_size, cur_tail = max_len, nums[idx]
        for i in range(idx, -1, -1):
            if cur_size == dp[i] and cur_tail % nums[i] == 0:
                res.append(nums[i])
                cur_size -= 1
                cur_tail = nums[i]

        return res[::-1]
```

```python
instance = Solution()
instance.largestDivisibleSubset([1, 2, 3])
```
