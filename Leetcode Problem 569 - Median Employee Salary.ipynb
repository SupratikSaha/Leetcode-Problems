# WINDOW FUNCTION

```python
WITH num_employees AS (
    SELECT
        company,
        COUNT(*) AS emp_count
    FROM 
        Employee
    GROUP BY
        company
),
employee_rank AS (
    SELECT 
        id, 
        company,
        salary,
        RANK() OVER (PARTITION BY company ORDER BY salary, id) AS ranked_salary
    FROM
        Employee
)
SELECT 
    e.id, 
    e.company,
    e.salary
FROM 
    employee_rank e
JOIN 
    num_employees n
ON
    e.company = n.company
AND 
    e.ranked_salary between n.emp_count / 2 AND n.emp_count / 2 + 1
```
